```python
packages = [
    ("Python", None),
    ("torch (PyTorch)", "torch"),
    ("dowhy (Causal)", "dowhy"),
    ("networkx", "networkx"),
    ("stable_baselines3 (RL)", "stable_baselines3")
]

import importlib, sys

print("Python version:", sys.version)
print("-"*40)

for name, module in packages[1:]:
    try:
        m = importlib.import_module(module)
        print(f"{name} ✅ INSTALLED")
    except:
        print(f"{name} ❌ NOT INSTALLED")
```
